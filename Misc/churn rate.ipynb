```python
def grouped_bar_chart_2(attrition_data, col1, col2,title =""):
    """
    Plot a bar chart for the categorical columns

    Arguments:
    dataframe name
    col1, col2  :categorical column name

    Output:
    Plot
    """
    temp =attrition_data.groupby([col1, col2]).size().to_frame().reset_index()
    temp.columns = [col1, col2,"frequency"]

    x = temp[col1].unique()
    data = []
    y = temp[temp[col2]==0].frequency
    data.append(go.Bar(name="no_attrition",
           x=x, y=y, text =y , textposition = "auto"))
    y = temp[temp[col2]==1].frequency
    data.append(go.Bar(name="attrition",
           x=x, y=y, text =y , textposition = "auto"))

    fig = go.Figure(data=data)
    # Change the bar mode
    fig.update_layout(barmode='group',
                      title = "{} and {} distribution".format(col1,col2))
    fig.show()
    

def plot_hist(dataframe, col_name):
    """Plot histogram"""
    data = [go.Histogram(x=dataframe[col_name],
                         marker=dict(
        color='#CC0E1D',  # Lava (#CC0E1D)
        #         color = 'rgb(200,0,0)'   # you can provide color in HEX format or rgb format, genrally programmers prefer HEX format as it is a single string value and easy to pass as a variable
    ))]
    layout = go.Layout(title="Histogram of {}".format(col_name))
    fig = go.Figure(data=data, layout=layout)
    return iplot(fig)


def grouped_bar_chart_1(attrition_data, col1, col2,title =""):
    """
    Plot a bar chart for the categorical columns

    Arguments:
    dataframe name
    col1, col2  :categorical column name

    Output:
    Plot
    """
    temp =attrition_data.groupby([col1, col2]).size().to_frame().reset_index()
    temp.columns = [col1, col2,"frequency"]


    data = []
    for _c in temp[col1].unique():
        x = temp[col2][temp[col1]==_c]
        y = temp[temp[col1] ==_c].frequency*100/temp[temp[col1] ==_c].frequency.sum()
        data.append(go.Bar(name=str(_c),
               x=x, y=y, text =y , textposition = "auto"))

    fig = go.Figure(data=data)
    # Change the bar mode
    fig.update_layout(barmode='group',
                      title = "{} and {} distribution".format(col1,col2))
    fig.show()

def plot_count_bar(dataframe_name, col_name, top_n=None):
    """
    Plot a bar chart for the categorical columns

    Arguments:
    dataframe name
    categorical column name

    Output:
    Plot
    """
    # create a table with value counts
    temp = dataframe_name[col_name].value_counts()
    if top_n is not None:
        temp = temp.head(top_n)
    # creating a Bar chart object of plotly
    data = [go.Bar(
            x=temp.index.astype(str),  # x axis values
            y=np.round(temp.values.astype(float) / temp.values.sum(), 4) * 100,  # y axis values
            text=['{}%'.format(i) for i in np.round(temp.values.astype(float) / temp.values.sum(), 4) * 100],
            # text to be displayed on the bar, we are doing this to display the '%' symbol along with the number on the bar
            textposition='auto',  # specify at which position on the bar the text should appear
            marker=dict(color='#0047AB'),)]  # change color of the bar
    # color used here Cobalt Blue
    fig = go.Figure(data=data)
    fig.update_layout(title  = "distribution of {} column".format(col_name))
    return iplot(fig)
```

```python
import matplotlib.pyplot as plt

def plot_confusion_matrix(data_path,experiment_name, cm, f1_score, classes,title='confusion matrix',
                          normalize=True,
                          cmap=plt.cm.Greens):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    cm_actual = cm.copy()
    if normalize:
        cm = (cm.astype('float') / cm.sum(axis=1)[:, np.newaxis])*100

    
    plt.figure()
    plt.imshow(cm, interpolation='nearest', cmap=cmap, aspect = 'equal')
    title = title + " F1 Score:" + str(np.round(f1_score,5)*100)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    
    plt.xticks(tick_marks, labels=[1,2,3,4,5], rotation=45)
    plt.yticks(tick_marks ,labels=[1,2,3,4,5], rotation=45)

    fmt_ = '.2f' 
    fmt =  'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt_),
                 horizontalalignment="center",
                 verticalalignment = "top",
                 color="white" if cm[i, j] > thresh else "black")
        plt.text(j, i, format(cm_actual[i, j], fmt),
                 horizontalalignment="center",
                 verticalalignment = "bottom",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('Actual Class')
    plt.xlabel('Predicted Class')
    plt.title(title)
    return plt

def fit_and_show_val(model,experiment_name,
                     train = [pd.DataFrame(),pd.DataFrame(), "train"],
                     test = [pd.DataFrame(),pd.DataFrame(), "test"],
                     display_model_name = "model",
                     weight_array = None,
                     title = 'confusion matrix for {} \n using {}',
                     train_model = False,
                     comment = ""
                    ):
    if train_model:
        if len(weight_array):
            model.fit(train[0], train[1], sample_weight =weight_array)
        else:
            model.fit(train[0], train[1])
    accs = {}
    # calculate and print accuracies
    acc_dict = {}
    acc_dict["comment"] = comment
    acc_dict["model_name"] = model.__class__.__name__
    acc_dict["accuracies"] = {}
    acc_dict["recalls"] = {}
    acc_dict["f1_scores"] = {}
    acc_dict["shapes"] = {}
    for t in [train,test]:
        print(f"running for {t[2]}")
        print(t[0].shape, t[1].shape)
        preds = model.predict(t[0])
        cm = confusion_matrix(t[1],preds)
        acc = accuracy_score(t[1],preds)
        recall = recall_score(t[1],preds,average='macro')
        f1_ = f1_score(t[1],preds,average='macro')
        accs[t[2]] = np.round(acc,4)
        acc_dict["accuracies"][t[2]] = acc*100
        acc_dict["recalls"][t[2]] = recall*100
        acc_dict["f1_scores"][t[2]] = f1_
        acc_dict["shapes"][t[2]] = t[0].shape, t[1].shape
        data_path = "Coding_Assignment/experiments"
        plot_confusion_matrix(data_path,experiment_name, cm, f1_,
                              classes = [1,2,3,4,5],title=title.format(t[2],display_model_name))
    

    
    pprint(accs)
    pprint(acc_dict)
    return acc_dict
```

```python
temp = pd.DataFrame(model_pred_info)
temp["train_accuracy"] = temp["accuracies"].apply(lambda x : x["train"])
temp["test_accuracy"] = temp["accuracies"].apply(lambda x : x["test"])
temp["train_recall"] = temp["recalls"].apply(lambda x : x["train"])
temp["test_recall"] = temp["recalls"].apply(lambda x : x["test"])
temp["train_f1_score"] = temp["f1_scores"].apply(lambda x : x["train"])
temp["test_f1_score"] = temp["f1_scores"].apply(lambda x : x["test"])

temp = temp.drop(["accuracies","recalls","f1_scores"],axis = 1)

display(temp.sort_values("test_recall",ascending=False ))
```

```python
def close_values(x,y, diff = 0.001, min_val= 0.75):
    """Condition = do_nothing acc close to replace_acc
acc > x"""
    if np.abs(x-y) < diff and (x > min_val) and (y>min_val):
        return True
    else:
        return False
```

# explanation:
Iterate over the dataframe with recall values, take the difference in train an test recall, find the model with least diffrence and display

```python
acceptable_differences = [0.001,0.005,0.01,0.05, 1,2,3,4,5,10,15,20,25]
for dif in acceptable_differences:
    temp["close"] = [close_values(i[0],i[1],diff=dif) for i in zip(temp.train_f1_score, 
                                                                                temp.test_f1_score)]
    best_model = temp[temp.close == True].sort_values("test_recall",
                                                                                       ascending = False)
    print(f"checking for a difference of {dif}")
    
    if best_model.shape[0] == 1:
        print(f"found best model")
        print("best_model")
        display(best_model)
        break
```

## Bayesian Opt with skopt

```python
y_train = y_train -1
y_test = y_test -1
```

```python
from skopt.space import Real, Integer
from skopt.utils import use_named_args
from skopt import gp_minimize
from skopt.plots import plot_convergence
import lightgbm as lgb


search_space = [Real(0.1, 0.9, name='learning_rate'),
                Integer(3, 15, name='max_depth'),
                Integer(6, 20, name='num_leaves'),
                Real(1, 5, name='scale_pos_weight'),
               Real(0.1, 0.9,name= 'feature_fraction'),
                Real(0.3,1, name = 'bagging_fraction'),
                Integer(20,100, 'max_bin'),
                Integer(20,100,'min_data_in_leaf'),
                Integer(0,100,'min_sum_hessian_in_leaf'),
                Real(0.01,1, name = 'subsample')
               ]

                                     
                                            
def BayesianOptimization(values):
    params = {
        'boosting_type': 'gbdt',
        'objective': 'multiclass',
        'num_class' : 5,
        'metric': 'auc_mu',
        'learning_rate': values[0],
        'max_depth': values[1],
        'nthread': -1,
        'num_leaves': values[2],
        'scale_pos_weight': values[3],
        'feature_fraction':values[4],
'bagging_fraction':values[5],
'max_bin':values[6],
'min_data_in_leaf':values[7],
'min_sum_hessian_in_leaf':values[8],
'subsample':values[9],
    }

    print('\nTesting next set of paramaters...', params)

    train_data = lgb.Dataset(X_train, label=y_train.astype(int))
    valid_data = lgb.Dataset(X_test, label=y_test.astype(int))
    lgbm = lgb.train(params, train_data, 10, valid_sets=valid_data, early_stopping_rounds=200,
                 verbose_eval=False)
    print(lgbm.predict(X_test).shape)
    neg_auc = round(-f1_score(y_test.astype(int), np.argmax(lgbm.predict(X_test.astype(int)),axis =1), average = 'macro'),9)

    print('f1_score: ', -neg_auc, ' of boosting iteration ', lgbm.current_iteration())
    return neg_auc
res_gp = gp_minimize(BayesianOptimization, search_space, random_state=26, n_jobs=-1, verbose=False, n_random_starts=50)
```
